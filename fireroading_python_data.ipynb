```python
# Create list of each class' data in the form of a json
courseDictList = []
for ind_class in set(relevant_courses):
    x = requests.get('https://fireroad.mit.edu/courses/lookup/' + ind_class)
    courseDictList.append(json.loads(x.text))
```

```python
# Create Dictionary of subject id to each metric for insertion into CSV
toPandas = {}
for i in courseDictList:
    if(i['subject_id'] not in toPandas):
        toPandas[i['subject_id']] = {}
    try:
        toPandas[i['subject_id']]['hours'] = i['out_of_class_hours'] + i['in_class_hours']
    except:
        if i['total_units'] <= 0:
            toPandas[i['subject_id']]['hours'] = 12
        else:
            toPandas[i['subject_id']]['hours'] = i['total_units']

    try:
        toPandas[i['subject_id']]['rating'] = i['rating']
    except:
        toPandas[i['subject_id']]['rating'] = 5.5
        
    toPandas[i['subject_id']]['units'] = i['total_units']
    toPandas[i['subject_id']]['offered_fall'] = i['offered_fall']
    toPandas[i['subject_id']]['offered_spring']= i['offered_spring']
#     try:
#         toPandas[i['subject_id']]['prerequisites']= i['prerequisites']
#     except:
#         pass
    try:
        toPandas[i['subject_id']]['prerequisites']= i['prerequisites']
    except:
        pass
    
    try:
        toPandas[i['subject_id']]['is_hass']= i['hass_attribute'] != ''
    except:
        toPandas[i['subject_id']]['is_hass']= False
    
    try:
        toPandas[i['subject_id']]['schedule']= i['schedule']
    except:
        pass
```

```python
# Construct Dictionary of class to 80 element array denoting which time periods are occupied by the class
days_dict = {"M":0, "T": 16, "W": 32, "R": 48, "F": 64}
class_to_schedule_list_dict = {}
offset = 18
for i in range(54):
    temp = [0] * 80
    if(not isinstance(df["schedule"][i], float)):
        for j in df["schedule"][i].split(";"):
            if(len(j) > 0):
                days_time = j.split(" ")
                days = days_time[0]
                time = days_time[1].split("-")
                for day in days:
                    if(len(time) == 0):
                        continue
                    if(float(time[0]) < 9):
                        start = float(time[0])*2 + 6
                    else:
                        start = float(time[0])*2 - 18
                    if len(time) == 1:
                        temp[days_dict[day] + int(start)] = 1
                        temp[days_dict[day] + int(start) + 1] = 1
                    else:
                        if(float(time[1]) < 9):
                            end = float(time[1])*2 + 6
                        else:
                            end = float(time[1])*2 - 18
                        time_slots = (end - start)
                        counter = 0
                        while(counter < time_slots):
                            temp[days_dict[day] + int(start) + counter] = 1
                            counter +=1
    
    class_to_schedule_list_dict[df["subject_id"][i]] = temp

print(class_to_schedule_list_dict)
```
